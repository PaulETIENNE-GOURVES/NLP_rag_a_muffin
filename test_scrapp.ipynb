```python
import requests
from bs4 import BeautifulSoup
url = "https://www.marmiton.org/recettes/recherche.aspx?aqt=muffin&page="

r = requests.get(url, allow_redirects=True)
```

```python
print(r.status_code)
```

```python
soup = BeautifulSoup(r.text, features="html.parser")
```

```python
title = soup.findAll(class_='card-content__title')
for elt in title:
    suffix = elt.get('href')
    print(suffix)
```
