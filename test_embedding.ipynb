```python
from llama_index.embeddings.ollama import OllamaEmbedding
```

```python
# Initialize the embedding model using Ollama
embed_model = OllamaEmbedding(
    model_name="mistral",
    base_url="http://localhost:11434",  # Default Ollama server URL
)

def recipe_to_embedding(recipe):
    # Combine relevant fields for embedding
    text = recipe["title"] + ", " + recipe["description"]
    return embed_model.get_text_embedding(text)
```

```python
from qdrant_client import QdrantClient
from qdrant_client.http import models
from llama_index.embeddings.ollama import OllamaEmbedding
```

```python
embed_model = OllamaEmbedding(
            model_name="mistral",
            base_url="http://localhost:11434",  # Default Ollama server URL
        )
```

```python
client = QdrantClient(path="./qdrant_data") 
```

```python
query = "Ingrdients : pommes"

query_embedding = embed_model.get_text_embedding(query)

results = client.search(
        collection_name="recipes",
        query_vector=query_embedding,
        limit=5
    )
```

```python
for result in results:
    print(f"ID: {result.id}, Score: {result.score}, Payload: {result.payload['description']}, Title: {result.payload['title']}")
```

```python
collections = client.get_collections()
print(collections)
```
